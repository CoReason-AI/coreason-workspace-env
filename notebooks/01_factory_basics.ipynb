## 2. Defining Your First Agent (Namespace and Taxonomy Consistency)

In this factory, every agent is strictly defined via a YAML manifest. 

> **CRITICAL RULE**: The `name` field in every `agent.yaml` MUST exactly match the `snake_case` name of the folder it resides in. This strictly ensures the internal Agent ID perfectly matches the namespace and folder routing logic.

Let's create a dummy agent folder and its YAML definition.

```python
import yaml
from pathlib import Path

agent_folder_name = "hello_world_agent"
agents_dir = Path(project_root) / "src" / "agents" / agent_folder_name
agents_dir.mkdir(parents=True, exist_ok=True)

agent_yaml_path = agents_dir / "agent.yaml"

# Define the agent manifest
agent_definition = {
    # TAXONOMY RULE: The name matches the folder exactly!
    "name": "hello_world_agent", 
    "description": "A simple deterministic sub-agent that says hello.",
    "orchestrator_type": "StateGraph",
    "capabilities": [
        "say_hello"
    ]
}

with open(agent_yaml_path, "w") as f:
    yaml.dump(agent_definition, f, sort_keys=False)

print(f"Successfully created agent definition at: {agent_yaml_path}")
```

## 3. Invoking the Factory Pipeline

Now that the definition exists, we use the **DeepAgents validation** pipeline to validate and compile it. 
The `factory_ceo` receives the request and delegates to:
1. `yaml_compiler`: Converts the YAML into LangGraph execution nodes.
2. `agent_validator`: Checks the generated nodes against strict rules (e.g., ensuring no legacy `AgentExecutor` or `LangSmith` calls are present).
